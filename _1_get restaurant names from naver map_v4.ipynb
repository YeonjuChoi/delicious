```python
# 네이버 지도 데이터 수집하기

from selenium import webdriver

##################################################
import time
##################################################

driver = webdriver.Chrome("./chromedriver/chromedriver.exe")
# 구버전 네이버지도 접속
driver.get("https://v4.map.naver.com")
# 네이버 지도 업데이트 후 안내메시지 끄기##########
driver.find_elements_by_css_selector("button.btn_close")[1].click()
##################################################

#3. 검색창에 검색어 입력하기 // 검색창: input#search-input
search_box = driver.find_element_by_css_selector("input#search-input")
search_box.send_keys("샤로수길 식당")
#4. 검색버튼 누르기 // 검색버튼: button.spm
search_button = driver.find_element_by_css_selector("button.spm")
search_button.click()

##################################################
# 지연시간
time.sleep(2)
##################################################

rest_name = []
rest_addr = []
stores = driver.find_elements_by_css_selector("div.lsnx")
for store in stores:
    # 세부 데이터 수집
    rest_name.append(store.find_element_by_css_selector("dt > a").text)
    rest_addr.append(store.find_element_by_css_selector("dd.addr").text)

for i in range(2,6):      
    sample = driver.find_element_by_xpath('//*[@id="panel"]/div[2]/div[1]/div[2]/div[2]/div/div/a['+str(i)+']')
    sample.send_keys('\n')
    time.sleep(2)
    stores = driver.find_elements_by_css_selector("div.lsnx")
    for store in stores:
        rest_name.append(store.find_element_by_css_selector("dt > a").text)
        rest_addr.append(store.find_element_by_css_selector("dd.addr").text)

while True:
    try:
        for i in range(3,7):      
            sample = driver.find_element_by_xpath('//*[@id="panel"]/div[2]/div[1]/div[2]/div[2]/div/div/a['+str(i)+']')
            sample.send_keys('\n')
            time.sleep(1)
            stores = driver.find_elements_by_css_selector("div.lsnx")
            for store in stores:
                rest_name.append(store.find_element_by_css_selector("dt > a").text)
                rest_addr.append(store.find_element_by_css_selector("dd.addr").text)
    except:
        break
driver.close()
print(rest_name)
print(rest_addr)
```
