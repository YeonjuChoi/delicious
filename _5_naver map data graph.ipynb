```python
#평점 그래프 그리기
none = 0
rate_under_4 = 0
rate_4_42 =0
rate_42_44 = 0
rate_44_46 = 0
rate_46_48 = 0
rate_48_5 = 0
rate_list = list(x_rest['rating'])
x_len = len(rate_list)
for x in rate_list:
    if x == "None":
        none+=1
    elif float(x) < 4:
        rate_under_4 +=1
    elif 4<=float(x)<4.2:
        rate_4_42 +=1
    elif 4.2<=float(x)<4.4:
        rate_42_44 +=1
    elif 4.4<=float(x)<4.6:
        rate_44_46 +=1
    elif 4.6<=float(x)<4.8:
        rate_46_48 +=1
    elif 4.8<=float(x):
        rate_48_5 +=1
x_rating_data = [round(rate_under_4/(x_len-none),3) ,round(rate_4_42/(x_len-none),3), round(rate_42_44/(x_len-none),3), round(rate_44_46/(x_len-none),3), round(rate_46_48/(x_len-none),3), round(rate_48_5/(x_len-none),3)]
x_rating_label=['under4', '4.0~4.2','4.2~4.4','4.4~4.6', '4.6~4.8','4.8~5']
x_rate_len = 0
x_rate_sum = 0
for x in rate_list:
    if x!='None':
        x_rate_len+=1
        x_rate_sum += float(x)
x_aver = round(x_rate_sum/x_rate_len,2)
        
```
