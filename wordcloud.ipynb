```
import csv
from collections import Counter
%matplotlib inline 
import matplotlib.pyplot as plt 
from wordcloud import WordCloud


# 한글 폰트 설치하기 (나눔 글꼴)
!apt-get install fonts-nanum* -qq

# 폰트매니저 불러오기/ 설치된 나눔글꼴 중 폰트고르기
import matplotlib.font_manager as fm
fontpath = '/usr/share/fonts/truetype/nanum/NanumGothicExtraBold.ttf'
font = fm.FontProperties(fname=fontpath, size=13)
# 기본 글꼴 변경
import matplotlib as mpl
mpl.font_manager._rebuild()
mpl.pyplot.rc('font', family='NanumBarunGothic')



a=open('/content/all.csv', encoding='cp949')
data=csv.reader(a)
b= list(data)

# print(b)

for i in range(int(len(b))):
    c=b[i]
    freq = []
    for tag in c:
        if tag not in ['error','']:
            freq.append(tag)
        else:
            pass

    tags=Counter(freq)
    print(tags.most_common(11))
    cloud_tags = WordCloud(font_path=fontpath, background_color='white', width=800, height=600) 
    cloud_tags = cloud_tags.fit_words(tags)
    plt.figure(figsize=(10,10)) 
    plt.axis('off') 
    plt.imshow(cloud_tags)  #데이터를 영상으로 출력
    plt.show()
```
